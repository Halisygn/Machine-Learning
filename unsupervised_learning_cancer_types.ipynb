```python
frames = [adeno, melanoma, meningioma,pilocytic]
df = pd.concat(frames,ignore_index=True)
df=df.drop( columns='Unnamed: 0')

# When I check the data first row was outlier. I will drop it.
df=df.drop([0])
types=df['type']
samples=df.drop( columns='type')
print(types)
```

# KMeans

```python
from sklearn.cluster import KMeans

ks = range(1, 11)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model=KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(samples)
    # Append the inertia to the list of inertias
    inertias.append( model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
# Create a KMeans model with 4 clusters
model = KMeans(n_clusters=4)

# Use fit_predict to fit model and obtain cluster labels
labels = model.fit_predict(samples)

# Create a DataFrame with labels and varieties as columns
df = pd.DataFrame({'labels': labels, 'varieties': types})

# Create crosstab
ct = pd.crosstab(df['labels'],df['varieties'])

# Display ct
print(ct)
```

```python
model = KMeans(n_clusters=4)
kmeans_features=model.fit_transform(samples)
sns.set(rc={'figure.figsize':(12,8)})
sns.scatterplot(kmeans_features[:,0], kmeans_features[:,1], hue=types, legend='full', palette=sns.color_palette("hls", 4))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer

# Create normalizer
normalize = Normalizer()
# Create KMeans 
kmeans = KMeans(n_clusters=4)

# Create pipeline
pipeline = make_pipeline(normalize,kmeans)

pipeline.fit_transform(samples)

# Calculate the cluster labels
labels = pipeline.predict(samples)

# Create a DataFrame with labels and species as columns: df
df = pd.DataFrame({'labels':labels,'varieties':types})

# Create crosstab as ct
ct = pd.crosstab(df['labels'],df['varieties'])

# Display ct
print(ct)
```

# Normalize KMeans

```python
# Create normalizer
normalize = Normalizer()
# Create KMeans instance
kmeans = KMeans(n_clusters=4)

# Create pipeline
pipeline = make_pipeline(normalize,kmeans)

kmeans_features=pipeline.fit_transform(samples)

# Scatter plot, coloring by variety_numbers
sns.scatterplot(kmeans_features[:,0], kmeans_features[:,1], hue=types, legend='full',palette=sns.color_palette("hls", 4))
plt.show()
```

# TSNE

```python
from sklearn.manifold import TSNE


# Create a TSNE instance: model
model = TSNE()

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(samples)

# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1st feature: ys
ys = tsne_features[:,1]

# Scatter plot, coloring by variety_numbers
sns.scatterplot(xs, ys, hue=types, legend='full', palette=sns.color_palette("hls", 4))
plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA

# Create a PCA model with 2 components: pca
pca = PCA(n_components=3)

# Fit the PCA instance to the scaled samples
pca.fit(samples)

# Transform the scaled samples: pca_features
pca_features = pca.transform(samples)

sns.scatterplot(pca_features[:,0], pca_features[:,1], hue=types, legend='full',palette=sns.color_palette("hls", 4))
plt.show()
```
